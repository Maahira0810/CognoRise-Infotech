```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score,recall_score,f1_score
from imblearn.over_sampling import SMOTE
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv('/content/creditcard.csv')
data
```

```python
x = data.drop('Class',axis=1)
y = data['Class']
```

```python
# Handling missing values of X
imputer_X = SimpleImputer(strategy = 'mean')
X_imputed =  imputer_X.fit_transform(x)
```

```python
# Handling missing values of y
imputer_y = SimpleImputer(strategy = 'most_frequent')
y_imputed =  imputer_y.fit_transform(y.values.reshape(-1,1))[:,0]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
```

**Addressing Class Imbalance**

```python
fraud_count = len(y_imputed[y_imputed == 1])
total_count = len(y_imputed)
print(f"Percentage of fraudulent transactions: {fraud_count/total_count:.2%}")
```

```python
smote = SMOTE(random_state = 42)
x_resampled, y_resampled =  smote.fit_resample(X_scaled, y_imputed)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size = 0.2, random_state = 42)
```

**Training Models**

```python
model_lr = LogisticRegression()
model_lr.fit(x_train, y_train)
```

```python
model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(x_train,y_train)
```

```python
# Evaluating Models
y_pred_lr = model_lr.predict(x_test)
precision_lr = precision_score(y_test, y_pred_lr)
recall_lr = recall_score(y_test, y_pred_lr)
f1_lr = f1_score(y_test, y_pred_lr)
```

```python
y_pred_rf = model_lr.predict(x_test)
precision_rf = precision_score(y_test, y_pred_lr)
recall_rf = recall_score(y_test, y_pred_lr)
f1_rf = f1_score(y_test, y_pred_lr)
```

```python
# Print Model Evaluation Metrics
print("\nLogistic Regression:")
print(f"Precision: {precision_lr:4f}")
print(f"Recall: {recall_lr:4f}")
print(f"F1-score: {f1_lr:4f}")
```

```python
print("\nRandom Forest:")
print(f"Precision: {precision_rf:4f}")
print(f"Recall: {recall_rf:4f}")
print(f"F1-score: {f1_rf:4f}")
```
